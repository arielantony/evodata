```python
filename = "tikiwiki_schema_history_new.txt"
```

```python
with open(filename, "r") as f:
    data = f.readlines()
```

```python
import re

revision = re.compile(r"^r(\d)+")
```

```python
matches = []

for line in data:
    l = line.strip('\n')
    if revision.match(l) is not None:
        print(revision.match(l)[0])
        matches.append(revision.match(l)[0])
        
matches = sorted(matches)
```
